```python
def print_unique_value(df):
    for column in df:
        if df[column].dtype=='object':
            print(f'{column}:{df[column].unique()}')
```

```python
df1.replace('No internet service','No',inplace=True)
df1.replace('No phone service','No',inplace=True)
```

```python
yes_no_column=['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
```

```python
for col in yes_no_column:
    df1[col].replace({'Yes':1,'No':0},inplace=True)
```

```python
for col in df1:
    print(f'{col}:{df1[col].unique()}')
```

```python
df1['gender'].replace({'Female':1,'Male':0},inplace=True)
```

```python
df2=pd.get_dummies(data=df1,columns=['InternetService'])
```

```python
df2.columns
```

```python
df2.head()
```

```python
df2.dtypes
```

```python
cols_to_scale=['tenure','MonthlyCharges','TotalCharges']
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
df2[cols_to_scale]=scaler.fit_transform(df2[cols_to_scale])
```

```python
df2.head()
```

```python
df2['InternetService_DSL'].replace({'True':1,'False':0},inplace=True)
df2['InternetService_Fiber optic'].replace({'True':1,'False':0},inplace=True)
df2['InternetService_No'].replace({'True':1,'False':0},inplace=True)
```

```python
df2.dtypes
```

```python
for col in df1:
    print(f'{col}:{df1[col].unique()}')
```

```python
df2['Contract'].replace({'Month-to-month':0,'One year':1,'Two year':2},inplace=True)
```

```python
df3=pd.get_dummies(data=df2,columns=['PaymentMethod']).astype(int)
```

```python
df3
```

```python
for col in df3:
    print(f'{col}:{df3[col].unique()}')
```

```python
x=df3.drop('Churn',axis='columns')
y=df3['Churn']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model=keras.Sequential([
    keras.layers.Dense(20,input_shape=(24,),activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])
model.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=5)
```

```python
model_hidden_layer=keras.Sequential([
    keras.layers.Dense(1000,input_shape=(24,),activation='relu'),
    keras.layers.Dense(100,activation='relu'),
    keras.layers.Dense(10,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])
model_hidden_layer.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['accuracy'])
```

```python
model_hidden_layer.fit(x_train,y_train,epochs=50)
```

```python
model_hidden_layer.evaluate(x_test,y_test)
```

```python
yp=model_hidden_layer.predict(x_test)
yp[:5]
```

```python
y_test[:5]
```

```python
y_pred=[]
for element in yp:
    if element>0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:5]
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(classification_report(y_test,y_pred))
```

```python
import seaborn as sns
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_pred)
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('predicted')
plt.ylabel('Truth')
```
