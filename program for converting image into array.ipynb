```python
from PIL import Image
import numpy as np

def image_to_numeric(image_path):
    # Open the image
    img = Image.open(image_path)

    # Convert image to grayscale
    img_gray = img.convert('P')
    img_gray.show()
    # Convert image to NumPy array
    img_array = np.array(img_gray)
    img_array
    return img_array

if __name__ == "__main__":
    # Replace 'your_image_path.jpg' with the actual path to your image file
    image_path = 'sanjana_image.png'

    # Convert image to numeric format
    numeric_array = image_to_numeric(image_path)

    # Print the resulting numeric array
    print("Numeric Array:")
    print(numeric_array)
    
```
